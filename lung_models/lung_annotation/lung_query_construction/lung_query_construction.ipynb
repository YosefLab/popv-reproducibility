```python
import anndata
import pandas as pd
import numpy as np
import scanpy as sc
```

```python
lca_10x = anndata.read('lca_10x.h5ad')
lca_ss2 = anndata.read('lca_ss2.h5ad')
lca = anndata.concat((lca_10x, lca_ss2))
lca.write('lca.h5ad')
```

```python
cols_to_keep = ['tissue', 'free_annotation', 'donor_id', 'sample', 'compartment', 'assay']
for i in lca.obs.columns:
    if i not in cols_to_keep:
        lca.obs.drop(i, axis=1, inplace=True)
del lca.obsm
lca
```

```python
lca.obs.rename(columns={"free_annotation": "paper_annotation"}, inplace=True)
```

```python
lca.obs['donor_assay'] = [i + '_' + j for i, j in zip(lca.obs['sample'], lca.obs['assay'])]
```
